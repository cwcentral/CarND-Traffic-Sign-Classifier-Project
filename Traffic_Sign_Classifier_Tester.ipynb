````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.
from sklearn.utils import shuffle
import cv2
import scipy.ndimage as ndimage
from scipy.misc import imread, imshow
from skimage.filters import rank
import skimage.morphology as morphology

print("***********************************")
print(" S4: Prepare Datasets")
print("***********************************")
print()

# Since we are looking at graphical features, aka in SIGNS 
# we can for go using RGB and speed up the model with gray
# GRAYSCALE, dim is 1 byte
depth_dim = 1

# helper functions
####################################################################
# convert to grasyscale
def get_gray(img):
    im = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    return np.resize(im[:,:], (32,32,1))

# convert to grasyscale with more contrast
def get_gray_eq(img):
    im = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    im = rank.equalize(im, selem=morphology.disk(25))
    return np.resize(im[:,:], (32,32,1))
    
# normalize pixel values +-1  
def get_norm(img):
    return (img / 255) 

# convert to normalized grayscale
def rgb_to_norm_gray(samples):
    tarr = []
    for img in samples:
        eqz_img = get_gray_eq(img)
        tarr.append(get_norm(eqz_img))
    return tarr
####################################################################

# enhance: convert to grayscale, add contrast, and normalize
if X_train[0].shape[2] == 3:   
    if depth_dim == 1:
        print("Current image data is COLOR")
        # convert datasets to grayscale
        print("convert training set")
        X_train = rgb_to_norm_gray(X_train)
        print("convert validation set")
        X_valid = rgb_to_norm_gray(X_valid)
        print("convert test set")
        X_test = rgb_to_norm_gray(X_test)
    else:
        X_train = get_norm(X_train)
        print("convert validation set")
        X_valid = get_norm(X_valid)
        print("convert test set")
        X_test = get_norm(X_test)

# randomize the datasets
X_train, y_train = shuffle(X_train, y_train)

# tmp shape variable (we tried 28x28 and 16x16 image kernels with little success)
image_shape_adj = X_train[0].shape
print("Current image data shape =", image_shape_adj)

# Since the datasets are not evenly distributed, let's make them!
# Increase the dataset aka augmentation

# a. Get bin counts
hist, edges = np.histogram(y_train, bins=n_classes)
print ('histogram =', hist)
print ('len =', len(hist))

#
# Originally the dataset showed low validation accuracy compared to the training data. 
# I saw lots of overfitting (validation accuracy vs training accuracy)
# that means the model matches to training set to well.

# We need to randomize the datasets more. One way is
# augmentation; increasing the dataset size with more random data.
#
# other methods could be changing 
# BATCH SIZE
# regularization
# more data
# changing the architecture (alternative to LeNet....)
#

print("***********************************")
print(" S4a: Augmenting data")
print("***********************************")
print()

# I want to evenly distribute new data upto 500 samples across all n_classes
max_sz = 500

# Find out how many samples need to be created to fulfill the max_sz
new_elem_sz = 0
for class_idx in range(0, n_classes):
    if hist[class_idx] < max_sz: 
        new_elem_sz += max_sz - hist[class_idx]
        
print("Augmentation will add new lements: " + str(new_elem_sz))   

# create placeholders
augmented_list_x = np.zeros((new_elem_sz, image_shape_adj[0], image_shape_adj[1], image_shape_adj[2]))
augmented_list_y = np.zeros(new_elem_sz)

# add images!
img_idx = 0

# make sure we loop through all classes without overfilling the temp space
while img_idx < new_elem_sz:       
    # if we reached our distribution requirement, this  will be false always.
    class_filled = False
    try:
        # loop thru classes
        for class_idx in range(0, n_classes):
            # if I don't have enough samples, add them
            while hist[class_idx] < max_sz: 
                
                # randomly find a correct sample, then...
                index = random.randint(0, len(X_train)-1)   #pick a random image in the dataset
                bin_idx = int(y_train[index])
                if bin_idx == class_idx:
                    class_filled = True
                    hist[bin_idx] += 1

                    # rotate it
                    rot_question = random.randint(0, 359)
                    rows,cols,dep = X_train[index].shape
                    M = cv2.getRotationMatrix2D((cols/2,rows/2),rot_question,1)
                    dst = cv2.warpAffine(X_train[index], M, (X_train[index].shape[0], 
                                                                            X_train[index].shape[1]))
                    # add noise
                    kern_val = random.randint(1, 5)
                    kernel = np.ones((kern_val,kern_val),np.float32)/25
                    dst = cv2.filter2D(dst,-1,kernel)

                    # add blur
                    blur_in = cv2.GaussianBlur(dst, (5,5), 20.0)
                    dst = cv2.addWeighted(dst, 2, blur_in, -1, 0)
                    dst = cv2.resize(dst,None,
                                      fx=random.uniform(1.0, 1.25), 
                                      fy=random.uniform(1.0, 1.25), 
                                      interpolation = cv2.INTER_CUBIC)[0:32, 0:32]
                    dst = dst[:,:,None]
                    
                     # add data to temp dataset
                    augmented_list_x[img_idx] = np.resize(dst, (32,32,depth_dim))
                    augmented_list_y[img_idx] = y_train[index]
                    img_idx += 1
                    
        if class_filled is False:  # we fulfilled the even distribution
            break
    except:
        print("reached max argument size. continuing....")
            

# add the new samples!
X_train = np.concatenate((X_train, augmented_list_x))        
y_train = np.concatenate((y_train, augmented_list_y))

print("New training dataset size: {}".format(len(X_train)))
    
 # let view of the new samples 
cell_w = 3
cell_h = 6
startIdx= n_train
tmp_img, X_img_arr = plt.subplots(cell_w, cell_h)
idx_row = 0
for idx in range(0,18):
    col = idx % cell_w
    # plot on relevant subplot
    target = startIdx+idx
    X_img_arr[col,idx_row].imshow(X_train[target].squeeze(), cmap="gray")
    if col == (cell_w - 1):
         # we have finished the current row, so increment row counter
        idx_row += 1

 # Let's show the corrected distributions
show_histogram(y_train, n_classes)        
````

### Model Architecture

````python
import tensorflow as tf
X_train, y_train = shuffle(X_train, y_train)

EPOCHS = 30
BATCH_SIZE = 96
local_learning_rate = 0.002
````

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
from tensorflow.contrib.layers import flatten

def LeNet(x):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    keep_prob = 0.8
    
    # SOLUTION: Layer 1: Convolutional. Input = 32x32xdepth_dim. Output = 28x28x6.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, depth_dim, 6), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(6))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)
 
    # SOLUTION: Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

    # SOLUTION: Layer 2: Convolutional. Output = 10x10x16.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(16))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)

    # SOLUTION: Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

    # SOLUTION: Flatten. Input = 5x5x16. Output = 400.
    fc0   = flatten(conv2)
    
    # SOLUTION: Layer 3: Fully Connected. Input = 400. Output = 120.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(120))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)
    if keep_prob != 0:
        fc1 = tf.nn.dropout(fc1, keep_prob)

    # SOLUTION: Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(120, 84), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(84))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)
    if keep_prob != 0:
        fc2 = tf.nn.dropout(fc2, keep_prob)

    # SOLUTION: Layer 5: Fully Connected. Input = 84. Output = n_classes.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(84, n_classes), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(n_classes))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
````

````python
# create batch holders
x = tf.placeholder(tf.float32, (None, 32, 32, depth_dim))
y = tf.placeholder(tf.int32, (None))
````

````python
# execute training
logits = LeNet(x)
one_hot_y = tf.one_hot(y, n_classes)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = local_learning_rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
#run model
validation_accuracy = []
training_accuracy = []

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    print("Training on {} examples...".format(num_examples))
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy.append(evaluate(X_valid, y_valid))
        training_accuracy.append(evaluate(X_train, y_train))
        print("EPOCH {} -- ".format(i+1) + \
              "Validation Accuracy = {:.3f}%".format(validation_accuracy[-1]*100) + \
              ", Training Accuracy = {:.3f}%".format(training_accuracy[-1]*100))
        
    saver.save(sess, './lenet')
    print("Model saved")
    
````

````python
plt.figure()    
plt.plot(validation_accuracy)
plt.title("Validation Accuracy (iterations)")
plt.figure()    
plt.plot(training_accuracy)
plt.title("Training Accuracy (iterations)")
plt.figure()   
t = np.linspace(2.0, 3.0, num=len(validation_accuracy))
plt.plot(t, training_accuracy)
plt.plot(t, validation_accuracy)
plt.title("Validation vs Training Accuracy (iterations)")
````
